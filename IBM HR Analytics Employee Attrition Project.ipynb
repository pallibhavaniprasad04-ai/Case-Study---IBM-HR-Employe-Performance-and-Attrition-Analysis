## 5. Bivariate Analysis: Explore relationships between two variables.

```python
#Is attrition higher in specific departments or roles? 
#Do lower salaries correlate with higher attrition? 
#Does overtime increase attrition risk?
```

```python
# 1️ Attrition by JobRole
plt.subplot()
sns.countplot(x='JobRole',  hue='Attrition' , data=df, palette='Set2')
plt.title('Attrition by Job Role')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2️ Attrition by Department
plt.subplot()
sns.countplot(x='Department', hue='Attrition', data=df, palette='cool')
plt.title('Attrition by Department')

plt.show
```

```python
# 3️ MonthlyIncome vs Attrition
plt.subplot()
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='pastel')
plt.title('Monthly Income vs Attrition')

plt.show()
```

```python
# 4️ OverTime vs Attrition
plt.figure(figsize=(5,4))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='Set3')
plt.title('OverTime vs Attrition')
plt.show
```

## Corr Findings

### OverTime vs Attrition

Employees with OverTime --> Yes have Higher attrition rate.

Employees are working overtime than regular times of company.

### MonthlyIncome vs Attrirtion
Lower MonthlyIncome has strongly linked to the attrition -- this will causes high loss to the company and HR need to check on this.



### Department vs Attrition

# 6.Multivariant Analysis:

### Does attrition differ across roles when considering income and satisfaction together?

```python
sns.catplot(x='JobSatisfaction', y='MonthlyIncome', hue='Attrition',
            kind='bar', data=df, height=5, aspect=1.5, palette='coolwarm')
plt.title('Income vs Job Satisfaction vs Attrition')
plt.show()
```

```python
#JobSatisfaction, MonthlyIncome, and Attrition
```

High job Satisfaction reduces attrition regardless of salary

## Insights and Summary:

### Based on HR Employee attrition analysis, the insights i have learned...

**Salary**- Low income employees will leave more, this will correlates with attrition and HR need to Review pay scales, especial for sales roles.

**OverTime**- Employees with frequent overtime are more likely to leave ,HR -->should monitor overtime hours and Implement better worklife balance policies.

**JobSatisfication**- Lower Satisfaction correlates with attrition and HR --> have to Conduct regular feedback sessions.

**Department**- Sales & HR have higher turnover and Focus retention strategies here...

**Age** - Younger employees (<30) show higher attrition .They need offer carrer growth programs.

## Business Impact:

HR can reduce attrition by 10-15% by focusing on work-life balance, Fair pay, and employee engagement initiatives.
